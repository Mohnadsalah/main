```python

# استيراد المكتبات الضرورية
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# قراءة بيانات الطلاب من ملف CSV
students_df = pd.read_csv("/content/study_performance.csv")

# عرض أول عشرة صفوف من البيانات
print("عرض أول عشرة صفوف:")
print(students_df.head(10))

# معلومات حول البيانات مثل عدد الصفوف وأسماء الأعمدة
print("\nمعلومات حول البيانات:")
print(students_df.info())

# تحليل توزيع درجات الرياضيات باستخدام الرسم البياني للتوزيع الكثافة النواة (KDE)
sns.displot(students_df['math_score'], kde=True, bins=20)
plt.title('توزيع درجات الرياضيات')

# تحليل العلاقة بين درجات القراءة والكتابة باستخدام مخطط مشترك (Jointplot)
sns.jointplot(x='reading_score', y='writing_score', data=students_df, kind='scatter')
plt.title('العلاقة بين درجات القراءة والكتابة')

# تحليل توزيع درجات الرياضيات بناءً على الجنس
sns.boxplot(x='gender', y='math_score', data=students_df)
plt.title('توزيع درجات الرياضيات بناءً على الجنس')

# عرض إحصائيات توصفية للبيانات
print("\nإحصائيات توصفية:")
print(students_df.describe())

# تحليل علاقة الدرجات بتحضير الطلاب للاختبار
sns.boxplot(x='test_preparation_course', y='math_score', data=students_df)
plt.title('علاقة تحضير الاختبار بدرجات الرياضيات')

plt.show()
```

```python
# تحليل توزيع درجات الرياضيات بناءً على العرق/الأصل العرقي
plt.figure(figsize=(10, 6))
sns.boxplot(x="race_ethnicity", y="math_score", data=students_df)
plt.title("توزيع درجات الرياضيات بناءً على العرق/الأصل العرقي")
plt.xticks(rotation=45)
plt.show()

# تحليل توزيع درجات القراءة والكتابة بناءً على مستوى التعليم الأبوي والعرق/الأصل العرقي
plt.figure(figsize=(12, 8))
sns.barplot(x="parental_level_of_education", y="reading_score", hue="race_ethnicity", data=students_df)
plt.title("توزيع درجات القراءة بناءً على مستوى التعليم الأبوي والعرق/الأصل العرقي")
plt.xticks(rotation=45)
plt.legend(title="العرق/الأصل العرقي")
plt.show()

plt.figure(figsize=(12, 8))
sns.barplot(x="parental_level_of_education", y="writing_score", hue="race_ethnicity", data=students_df)
plt.title("توزيع درجات الكتابة بناءً على مستوى التعليم الأبوي والعرق/الأصل العرقي")
plt.xticks(rotation=45)
plt.legend(title="العرق/الأصل العرقي")
plt.show()

# تحليل توزيع درجات الرياضيات بناءً على مستوى التعليم الأبوي والعرق/الأصل العرقي ونوع الغداء
plt.figure(figsize=(12, 8))
sns.barplot(x="parental_level_of_education", y="math_score", hue="race_ethnicity", data=students_df, ci=None)
plt.title("توزيع درجات الرياضيات بناءً على مستوى التعليم الأبوي والعرق/الأصل العرقي ونوع الغداء")
plt.xticks(rotation=45)
plt.legend(title="العرق/الأصل العرقي")
plt.show()
```

```python
# تحليل علاقة مستوى التعليم الأبوي بدرجات الرياضيات
plt.figure(figsize=(10, 6))
sns.boxplot(x="parental_level_of_education", y="math_score", data=students_df)
plt.title("علاقة مستوى التعليم الأبوي بدرجات الرياضيات")
plt.xticks(rotation=45)
plt.show()

# تحليل توزيع درجات القراءة بناءً على نوع الغداء
plt.figure(figsize=(8, 6))
sns.violinplot(x="lunch", y="reading_score", data=students_df)
plt.title("توزيع درجات القراءة بناءً على نوع الغداء")
plt.show()

# تحليل توزيع درجات الكتابة بناءً على استعداد الطلاب للاختبار
plt.figure(figsize=(8, 6))
sns.boxplot(x="test_preparation_course", y="writing_score", data=students_df)
plt.title("توزيع درجات الكتابة بناءً على استعداد الطلاب للاختبار")
plt.show()

# تحليل توزيع درجات الرياضيات بناءً على مستوى التعليم الأبوي ونوع الغداء
plt.figure(figsize=(12, 8))
sns.barplot(x="parental_level_of_education", y="math_score", hue="lunch", data=students_df)
plt.title("توزيع درجات الرياضيات بناءً على مستوى التعليم الأبوي ونوع الغداء")
plt.xticks(rotation=45)
plt.legend(title="نوع الغداء")
plt.show()
```
